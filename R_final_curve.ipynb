```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def exponential_curve(t, alpha, q0):
    return q0*np.exp(-alpha*t)

# Generate x values
x_values = np.linspace(1, 50, 5)

# Set the initial production rate (q0)
q0 = 50000

# Calculate corresponding y values using the function
y_values = exponential_curve(x_values, 1.7, q0)

# Plot the curve
plt.plot(x_values, y_values, label=r'$qi = q0 \cdot e^x$')
plt.title('Exponential Curve: $qi = q0 \cdot e^x$')
plt.xlabel('x')
plt.ylabel('qi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def exponential_np(qi, Di, t):
    if Di == 0:
        return qi * t
    else:
        return qi / Di * (1 - np.exp(-Di * t))
y_values = exponential_np(q0, 0.05, x_values)

# Plot the curve
plt.plot(x_values, y_values, label=r'$qi = q0 \cdot e^x$')
plt.title('Exponential Curve: $qi = q0 \cdot e^x$')
plt.xlabel('x')
plt.ylabel('qi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation
def custom_equation(t, b, Di, qi):
    return qi / ((1 - b) * Di) * (1 - (1 + b * Di * t) ** (1 - 1 / b))

# Generate values for t
t_values = np.linspace(0, 10, 100)  # Adjust the range and number of points as needed

# Set the parameters
b_value = 0.5  # Replace with your desired value
Di_value = 1.0  # Replace with your desired value
qi_value = 100.0  # Replace with your desired value

# Calculate the corresponding y values using the equation
y_values = custom_equation(t_values, b_value, Di_value, qi_value)

# Plot the graph
plt.plot(t_values, y_values, label='Custom Equation')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Graph of the Custom Equation')
plt.legend()
plt.grid(True)
plt.show()
```
